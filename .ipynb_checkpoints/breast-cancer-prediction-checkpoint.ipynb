# Project

##### Import all needed libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from datetime import time
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

##### Read in the csv file from the data directory and make a copy to work from.

```python
cancer_csv = pd.read_csv("data\cancer.csv")
cancer_csv_copy = cancer_csv
```

##### Drop the ID column as it does not assist with the fitting and training the model.

```python
cancer_csv_copy.drop("id", axis=1, inplace=True)
```

```python
cancer_csv_validation_patients = cancer_csv_copy.tail(20)
```

```python
cancer_csv_copy.drop(cancer_csv_copy.tail(20).index, inplace=True)
```

```python
cancer_csv_copy
```

```python
cancer_csv_validation_patients
```

```python
X = cancer_csv_copy.drop('diagnosis', axis=1)

y = cancer_csv_copy['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = RandomForestClassifier(n_estimators=1000)
clf = clf.fit(X_train, y_train)

y_pred_test = clf.predict(X_test)
accuracy_score(y_test, y_pred_test)
#clf.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred_test))
```
